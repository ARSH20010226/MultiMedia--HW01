```python
def horizontal_flip(Input_image) : 
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)


  for y in range(height):
    # get the left and right (mirror) pixels
    for x in range(width//2):
      new_image[y,width-1-x]=Input_image[y,x]
      new_image[y,x]=Input_image[y,width-1-x]
  return new_image
```

```python
def transpose1(Input_image) : 
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)

  #Define paramaters of transpose
  angle = math.pi / 2  # angle in radian
  center_x = width / 2
  center_y = height / 2

  #Create a new image that is rotated to the right of our input image
  for x in range(width):
    for y in range(height):
      # Compute coordinate in input image
      xp = int((x - center_x) * math.cos(angle) - (y - center_y) * math.sin(angle) + center_x)
      yp = int((x - center_x) * math.sin(angle) + (y - center_y) * math.cos(angle) + center_y)
      if 0 <= xp < width and 0 <= yp < height:
        new_image[y,x]=Input_image[yp,xp]
  return new_image
```

```python
#Test transpose1 function
PartB1 = transpose1(Original_image)

#Show results
fig = plt.figure(figsize=(10, 7))
rows=2 
columns=2

#Show an original(input) image
fig.add_subplot(rows, columns, 1)
plt.imshow(Original_image,cmap="gray")
plt.axis("on")
plt.title("Original image")

#Show the transposed to right image 
fig.add_subplot(rows, columns, 2)
plt.imshow(PartB1,cmap="gray")
plt.axis("on")
plt.title("Left transposed image:")
```

```python
def Right_shift(Input_image) : 
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)
  
  #new_image=np.roll(Input_image,2)
  for y in range(height):
    for x in range(width):
      if x<width-2 :
        new_image[y,x+2]=Input_image[y,x]
      elif x==width-2 :
        new_image[y,0]=Input_image[y,x]
      elif x==width-1 : 
        new_image[y,1]=Input_image[y,x]
  return new_image
```

```python
def Left_shift(Input_image) :   
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)
  
  #Make shift with inbuilt functions and numpy properties
  #new_image=np.roll(Input_image,-2)

  for y in range(height):
    for x in range(width):
      if x>1 : 
        new_image[y,x-2]=Input_image[y,x]
      elif x==0 : 
        new_image[y,width-2]=Input_image[y,x]      
      elif x==1 :
        new_image[y,width-1]=Input_image[y,x]
  
  return new_image
```

```python
def Up_shift(Input_image) : 
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)
  
  #new_image=np.roll(Input_image,2,axis=0)
  
  for x in range(width):    
    for y in range(height):
      if y>1 :
        new_image[y-2,x]=Input_image[y,x] 
      elif y==0 : 
        new_image[height-2,x]=Input_image[y,x] 
      elif y==1 :
        new_image[height-1,x]=Input_image[y,x]
  return new_image
```

```python
def Down_shift(Input_image) : 
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)
  
  
  #new_image=np.roll(Input_image,-2,axis=0)

  for x in range(width):
    for y in range(height):
      if y<height-2 : 
        new_image[y+2,x]=Input_image[y,x]
      elif y==height-2 : 
        new_image[1,x]=Input_image[y,x]
      elif y==height-1 :
        new_image[0,x]=Input_image[y,x]
  return new_image
```

```python
def HW1_MSE(image1 , image2) :
  image1 = np.asarray(image1)
  image2 = np.asarray(image2)
  # the 'Mean Squared Error' between the two images is the
  # sum of the squared difference between the two images;
  # NOTE: the two images must have the same dimension
  err = np.sum((image1.astype("float") - image2.astype("float")) ** 2)
  err /= float(image1.shape[0] * image1.shape[1])
  # return the MSE, the lower the error, the more "similar"
  # the two images are
  return err
```

```python
def Q3_1(Input_image):
  #Get height and width of an image
  height,width=Input_image.shape
  #Make a new image like our Input image and fills it with zeros
  new_image = np.zeros_like(Input_image)

  #Do the decreasing 20 units from each pixels with using two for loops -> One of them on a height of an image and another one on a width of an image 
  for x in range(height): 
    for y in range(width): 
      if Input_image[x,y]-20>=0: 
        new_image[x,y]=float(Input_image[x,y] -20)
      elif Input_image[x,y]-20<0: 
        new_image[x,y] = 0 
  return new_image
```

```python
def Q3_2(Input_image):
  #Create a list from our images
  images =[] 
  #Append an original(input) image to our list
  images.append(Input_image)
  #Append a right shifted image to our list 
  images.append(Right_shift(Input_image))
  #Append a left shifted image to our list
  images.append(Left_shift(Input_image))

  #Assign first element of list to our an average image
  average_image = images[0]
  #Calculate average of images usign a for loop on length of our list 
  for i in range(len(images)):
    if i == 0:
      pass
    else:
      alpha = 1.0/(i + 1)
      beta = 1.0 - alpha
      average_image = cv2.addWeighted(images[i], alpha, average_image, beta, 0.0)
  return average_image
```
